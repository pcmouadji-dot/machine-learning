```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as skl
from sklearn.ensemble import RandomForestClassifier
```

```python
train_data=pd.read_csv("train.csv")
train_data=train_data.drop(columns=["Name"])
train_data=pd.get_dummies(train_data,dtype=int,prefix="HomePlanet",columns=["HomePlanet"])
train_data["VIP"]=train_data["VIP"].map({False:0,True:1}).fillna(-1).astype(int)

#train_data["VIP"]=train_data["VIP"].astype(int)
train_data["Destination"]=train_data["Destination"].fillna("TRAPPIST-1e")
train_data=pd.get_dummies(train_data,dtype=int,prefix="Destination",columns=["Destination"])
train_data['Transported']=train_data['Transported'].astype(int)
id_split=train_data["PassengerId"].str.split("_",expand=True)
train_data['group']=id_split[0].astype(int)
train_data['num_grp']=id_split[1].astype(int)
train_data=train_data.drop(columns="PassengerId")
#train_data['CryoSleep']=train_data['CryoSleep'].astype(int)
mode=train_data['CryoSleep'].mode()[0]
train_data['CryoSleep']=train_data['CryoSleep'].fillna(mode)
train_data['CryoSleep']=train_data['CryoSleep'].astype(int)
train_data['RoomService']=train_data['RoomService'].fillna(train_data['RoomService'].median())
train_data['FoodCourt']=train_data['FoodCourt'].fillna(train_data['FoodCourt'].median())
train_data['VRDeck']=train_data['VRDeck'].fillna(train_data['VRDeck'].median())
train_data['Spa']=train_data['Spa'].fillna(train_data['Spa'].median())
train_data['ShoppingMall']=train_data['ShoppingMall'].fillna(train_data['ShoppingMall'].median())

train_data['spend']=train_data['ShoppingMall']+train_data['Spa']+train_data['VRDeck']+train_data['FoodCourt']+train_data['RoomService']
train_data=train_data.drop(columns=['ShoppingMall','Spa','VRDeck','FoodCourt','RoomService'])

train_data['Age']=train_data['Age'].fillna(train_data['Age'].median())
#print(train_data["Transported"].isnull().sum())
train_data['Cabin']=train_data['Cabin'].fillna('U/-1/U')
place=train_data['Cabin'].str.split("/",expand=True)
train_data['cabin_deck']=place[0]
train_data['cabin_num']=place[1].astype(int)
train_data['cabin_side']=place[2]
train_data=pd.get_dummies(train_data,prefix='cabin_side',columns=['cabin_deck','cabin_side'],dtype=int)
train_data=train_data.drop(columns='Cabin')
x=train_data.drop(columns='Transported')
y=train_data['Transported']
```

```python
model=RandomForestClassifier()
model.fit(x,y)
```

**cleaning test data**

```python
test_data = pd.read_csv("test.csv")
passenger_ids = test_data["PassengerId"]

test_data = test_data.drop(columns=["Name"])
test_data = pd.get_dummies(test_data, dtype=int, prefix="HomePlanet", columns=["HomePlanet"])
test_data["VIP"] = test_data["VIP"].map({False: 0, True: 1}).fillna(-1).astype(int)

test_data["Destination"] = test_data["Destination"].fillna("TRAPPIST-1e")
test_data = pd.get_dummies(test_data, dtype=int, prefix="Destination", columns=["Destination"])

id_split = test_data["PassengerId"].str.split("_", expand=True)
test_data['group'] = id_split[0].astype(int)
test_data['num_grp'] = id_split[1].astype(int)
test_data = test_data.drop(columns="PassengerId")

mode = test_data['CryoSleep'].mode()[0]
test_data['CryoSleep'] = test_data['CryoSleep'].fillna(mode)
test_data['CryoSleep'] = test_data['CryoSleep'].astype(int)

test_data['RoomService'] = test_data['RoomService'].fillna(test_data['RoomService'].median())
test_data['FoodCourt'] = test_data['FoodCourt'].fillna(test_data['FoodCourt'].median())
test_data['VRDeck'] = test_data['VRDeck'].fillna(test_data['VRDeck'].median())
test_data['Spa'] = test_data['Spa'].fillna(test_data['Spa'].median())
test_data['ShoppingMall'] = test_data['ShoppingMall'].fillna(test_data['ShoppingMall'].median())

test_data['spend'] = test_data['ShoppingMall'] + test_data['Spa'] + test_data['VRDeck'] + test_data['FoodCourt'] + test_data['RoomService']
test_data = test_data.drop(columns=['ShoppingMall', 'Spa', 'VRDeck', 'FoodCourt', 'RoomService'])

test_data['Age'] = test_data['Age'].fillna(test_data['Age'].median())

test_data['Cabin'] = test_data['Cabin'].fillna('U/-1/U')
place = test_data['Cabin'].str.split("/", expand=True)
test_data['cabin_deck'] = place[0]
test_data['cabin_num'] = place[1].astype(int)
test_data['cabin_side'] = place[2]
test_data = pd.get_dummies(test_data, prefix='cabin_side', columns=['cabin_deck', 'cabin_side'], dtype=int)
test_data = test_data.drop(columns='Cabin')

```

```python
predictions=model.predict(test_data)
submission=pd.DataFrame({"PassengerId":passenger_ids,"Transported":predictions.astype(bool)})
submission.to_csv("submission.csv",index=False)
```
