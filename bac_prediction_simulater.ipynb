```python


model=RandomForestRegressor(n_estimators=100)
Y=df["bac_exam_grade"]
X=df.drop(columns="bac_exam_grade")
trainx,testx,trainy,testy=train_test_split(X,Y,test_size=0.1)
model.fit(trainx,trainy)
model.score(testx,testy)
```
