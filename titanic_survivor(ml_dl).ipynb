```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn as skl
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow import keras

df=pd.read_csv("train.csv")
#print(df["cabin"].isnull().sum()) #most the cabin values are NaN so we drop this colum
#i don't think that the id will effect if you would survive
df['Is_Master'] = df['Name'].str.contains('Master', na=False).astype(int)
df['Is_Mr'] = df['Name'].str.contains('Mr\.', na=False).astype(int)
df['Is_Mrs'] = df['Name'].str.contains('Mrs\.', na=False).astype(int)
df['Is_Miss'] = df['Name'].str.contains('Miss\.', na=False).astype(int)

df=df.drop(columns=["Cabin","PassengerId","Ticket","Name"])
age_median=df["Age"].median()
df["Age"]=df["Age"].fillna(age_median)
#print(df["Embarked"].isnull().sum()) so we drop this two rows
df=df.dropna(subset=["Embarked"])
df["Sex"]=df["Sex"].replace({"male":1,"female":0})

df=pd.get_dummies(df,columns=["Embarked"],dtype=int,prefix="Embarked")


x=df.drop(columns="Survived")
y=df["Survived"]
"""model=GradientBoostingClassifier(n_estimators=100,learning_rate=0.1)

model.fit(x,y)"""
scale=StandardScaler()
x=scale.fit_transform(x)
model=keras.Sequential([
    keras.layers.Dense(128,input_shape=(13,),activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64,activation="relu"),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(32,activation="relu"),
    keras.layers.Dropout(0.2),

    keras.layers.Dense(10,activation="tanh"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1,activation="sigmoid"),
])
display(x)
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
model.fit(x,y,epochs=40)



tdf=pd.read_csv("test.csv")
passenger_ids = tdf["PassengerId"]

tdf['Is_Master'] = tdf['Name'].str.contains('Master', na=False).astype(int)
tdf['Is_Mr'] = tdf['Name'].str.contains('Mr\.', na=False).astype(int)
tdf['Is_Mrs'] = tdf['Name'].str.contains('Mrs\.', na=False).astype(int)
tdf['Is_Miss'] = tdf['Name'].str.contains('Miss\.', na=False).astype(int)
tdf=tdf.drop(columns=["Cabin","PassengerId","Ticket","Name"])
age_median=tdf["Age"].median()
tdf["Age"]=tdf["Age"].fillna(age_median)
tdf["Fare"] = tdf["Fare"].fillna(tdf["Fare"].median())
tdf["Embarked"] = tdf["Embarked"].fillna("S") # Fill with the most common port
tdf["Sex"]=tdf["Sex"].replace({"male":1,"female":0})

tdf=pd.get_dummies(tdf,columns=["Embarked"],dtype=int,prefix="Embarked")

tdf=tdf[tdf.columns]
tdf=scale.transform(tdf)
prediction=model.predict(tdf)
prediction=np.round(prediction).flatten().astype(int)
submission=pd.DataFrame({"PassengerId":passenger_ids,"Survived":prediction})
submission.to_csv("submission.csv",index=False)
```
